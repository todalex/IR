```python
import requests
from bs4 import BeautifulSoup
from newspaper import Article
from tqdm import tqdm
import pandas as pd
```

```python
def repetetive(links, url):
    for link in links:
        if link.a['href'] in url:
            return True
    return False
```

```python
def scrap_year(year: int):
    page=1
    scraped_data = []
    url_list = []
    while True:
        page += 1
        page_url = f"https://www.tasnimnews.com/fa/latest-news?date={year}%2f01%2f01&sub=-1&service=-1&page={page}"
        html = requests.get(page_url).text
        soup = BeautifulSoup(html, features='lxml')
        links = soup.find_all('article')
        #print(links)
        #if repetetive(links, url_list):
        if page ==5:
            break
        for link in tqdm(links):
            news_url = 'https://www.tasnimnews.com' + link.a['href']
            url_list.append(link.a['href'])
            try:
                article = Article(news_url)
                article.download()
                article.parse()
                scraped_data.append({'url':news_url, 'text':article.text, 'title':article.title})
            except:
                print(f'faile :{news_url}')
    print(len(scraped_data))
    df=pd.DataFrame(scraped_data)
    df.to_csv(f'tasnimnews-{year}.csv')
```

# Query processing¶

```python
with open('hamshahri.json', 'r') as f:
  corpus = json.load(f)
type(corpus), len(corpus), corpus[0]
```

```python
docs = [d['summary'] for d in corpus]
# tf-idf
vectorizer = TfidfVectorizer()
tfidf_docs = vectorizer.fit_transform(docs)
tfidf_docs.shape, len(vectorizer.vocabulary_)
```

```python
list(vectorizer.vocabulary_.keys())[:10]
# query
query = 'جرایم رانندگی'

tfidf_query = vectorizer.transform([query])[0]
```

```python
# similarities
cosines = []
for d in tqdm(tfidf_docs):
    cosines.append(float(cosine_similarity(d, tfidf_query)))
```

```python
"رانندگی" in docs["text"][246].split()
```

```python
# sorting
k = 10
sorted_ids = np.argsort(cosines)
sorted_ids
for i in range(k):
    cur_id = sorted_ids[-i-1]
    print(cur_id)
    print(docs['text'][cur_id], cosines[cur_id])
```
